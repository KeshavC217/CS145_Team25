```python
import pandas as pd
```

```python
data = pd.read_csv("data/train.csv")
```

One hot encoding:

```python
data = pd.get_dummies(data, prefix='State', columns=['Province_State'])

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

data.head()
```

Cleaning date formatting

```python
day = []
month = []
year = []

for date in data['Date']:
    day = date[0:2]
    month = date[3:5]
    year = date[7:]

data['Day'] = day
data['Month'] = month
data['Year'] = year
data.drop(columns=['Date'], inplace=True)

data.head()
```

```python
data.to_csv("data/cleaned_data.csv")
```
